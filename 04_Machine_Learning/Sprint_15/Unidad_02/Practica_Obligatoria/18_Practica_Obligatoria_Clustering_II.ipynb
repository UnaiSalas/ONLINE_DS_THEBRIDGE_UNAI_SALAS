```python
features = ["Taurina", "Cafena"]
all_features = df.columns.to_list()
```

```python
link_method = linkage(df[features], method = 'ward')
plt.figure(figsize =(15,6))
dend = dendrogram(link_method)
```

```python
clustering_model = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
clustering_model.fit(df[features])

data_labels = clustering_model.labels_
sns.scatterplot(x='Cafena',      
                y='Taurina', 
                data=df, 
                hue=data_labels, palette = "tab10").set_title('Cafena/Taurina');
```

```python
data_labels = clustering_model.labels_
sns.scatterplot(x='Cafena',      
                y='cido Ctrico', 
                data=df, 
                hue=data_labels, palette = "tab10").set_title('Cafena/cido Ctrico');
```

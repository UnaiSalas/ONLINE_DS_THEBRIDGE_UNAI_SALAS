### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
#plt.plot(clean_data.city, clean_data.population_total)
clean_data_sorted = clean_data.sort_values(by='population_total', ascending=False)
clean_data_sorted["population_total"] = clean_data_sorted["population_total"]/1000

colores = ['royalblue', 'lightgrey', 'salmon', 'olive', 'darksalmon']
# Seleccionar las primeras 5 filas
top_5_cities = clean_data_sorted.head(5)

bars = plt.bar(top_5_cities['city'], top_5_cities['population_total'], label='Population', color =colores)

for bar in bars:
    poblacion = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, poblacion, f'{poblacion:,}',
             ha = 'center', va = 'bottom', color = 'black', fontsize = 10)
    
plt.xlabel('Ciudades')  
plt.ylabel('Habitantes (miles)') 
plt.yticks(np.arange(0, 5001, 1000))
plt.show()
```

```python
plt.hist(clean_data['area_total_km2'], bins = 50, color = 'steelblue', alpha = 0.7)
plt.xlabel('superficie (km2)', fontname = "Bradley Hand itc", fontsize = 18)
plt.ylabel('Numero de ciudades', fontname = "Bradley Hand itc", fontsize = 18)
plt.legend(['Hist Area'])
```

```python
clean_data["population_total"] = clean_data["population_total"]/1000
plt.scatter(clean_data["area_water_km2"], clean_data["population_total"], alpha = 0.3, color = 'g')

plt.xlabel("Area de agua en la ciudad (km2)")
plt.ylabel("Poblacion (miles habitantes)")
plt.title('Poblacion por kilometro cuadrado de agua')
plt.yticks(np.arange(0, 4001, 500))

props_flecha = {
    "arrowstyle" : "->"
}
plt.annotate("Outliers", xy = (1300, 3900000), xytext = (800, 3500000), arrowprops = props_flecha)
plt.annotate("Outliers", xy = (965, 1360000), xytext = (800, 3500000), arrowprops = props_flecha)
```

```python
plt.boxplot(clean_data["elevation_m"],
            vert = False,
            patch_artist = True,
            boxprops = dict(facecolor = "red", color = "black"),
            medianprops = dict(color = "black")
            )
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)")
```

```python
fig, axs = plt.subplots(2, 2, figsize = (12, 10))

# Barra de las ciudades ms pobladas
df_top5 = clean_data.sort_values(by = 'population_total', ascending = False).head(5)
colores = ['mediumslateblue', 'silver', 'coral', 'olive', 'indianred']
bars = axs[0, 0].bar(df_top5['city'], df_top5['population_total'], color=colores, alpha = 0.7)
axs[0, 0].set_xlabel('Ciudades')
axs[0, 0].set_ylabel('Habitantes')
axs[0, 0].set_title('Ciudades ms pobladas de California')

for bar in bars:
    poblacion = bar.get_height()
    axs[0, 0].text(bar.get_x() + bar.get_width() / 2, poblacion, f'{poblacion:,}',
                   ha='center', va='bottom', color='black', fontsize=10)

# Histograma de reas totales
axs[0, 1].hist(clean_data['area_total_km2'], bins=50, color='steelblue', alpha = 0.7)
axs[0, 1].set_xlabel('Superficie (km2)', fontname="Bradley Hand itc", fontsize = 18)
axs[0, 1].set_ylabel('Numero de ciudades', fontname="Bradley Hand itc", fontsize = 18)
axs[0, 1].legend(['Hist Area'])

# Dispersin de poblacin por rea de agua
axs[1, 0].scatter(clean_data["area_water_km2"], clean_data["population_total"], alpha = 0.3, color = 'g')
axs[1, 0].set_xlabel("Area de agua en la ciudad (km2)")
axs[1, 0].set_ylabel("Poblacion (miles habitantes)")
axs[1, 0].set_title('Poblacion por kilometro cuadrado de agua')
axs[1, 0].annotate("Outliers", xy = (1300, 3900000), xytext = (800, 3500000), arrowprops = props_flecha)
axs[1, 0].annotate("Outliers", xy = (965, 1360000), xytext = (800, 3500000), arrowprops = props_flecha)

# Diagrama de caja de la altitud
axs[1, 1].boxplot(clean_data["elevation_m"],
            vert = False,
            patch_artist = True,
            boxprops = dict(facecolor = "red", color = "black"),
            medianprops = dict(color = "black")
            )
axs[1, 1].set_title('Diagrama de caja de la Altitud de las ciudades de California')
axs[1, 1].set_xlabel('Altitud (m)')
```

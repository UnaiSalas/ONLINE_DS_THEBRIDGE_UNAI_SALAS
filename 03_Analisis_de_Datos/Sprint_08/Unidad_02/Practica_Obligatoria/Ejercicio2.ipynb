```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from bootcampviztools import *
```

```python
df_titanic = pd.read_csv("./data/titanic.csv")
```

```python
df_titanic.drop(["deck","parch", "embarked"], axis = 1, inplace = True)
```

```python
#Tasa de mortalidad del Titanic
sns.countplot(df_titanic, x="alive")

plt.title("Tasa de supervivencia")
plt.xlabel("Superviviente")
plt.ylabel("Cantidad Pasajeros")

plt.savefig("t_mortalidad.png")
```

```python
total_pasajeros = len(df_titanic)
porcentaje_supervivencia = df_titanic['alive'].value_counts(normalize=True) * 100
sns.countplot(df_titanic, x="alive", hue="alive")

for index, value in enumerate(porcentaje_supervivencia):
    plt.text(index, value, f'{value:.2f}%', ha='center')

plt.title("Tasa de supervivencia")
plt.xlabel("Superviviente")
plt.ylabel("Cantidad Pasajeros")

```

```python
# 2- En esa ocasin los datos apoyan que se dio el "Las mujeres y los nios primero", pero...
plot_categorical_relationship_fin(df_titanic, "who","alive", relative_freq= True, show_values= True)

plt.savefig("grupos_personas_vivas.png")
```

```python
x,y = 'class', 'survived'

df1 = df_titanic.groupby(x)[y].value_counts(normalize=True)
df1 = df1.mul(100)
df1 = df1.rename('percent').reset_index()

g = sns.catplot(x=x,y='percent',hue=y,kind='bar',data=df1)
g.ax.set_ylim(0,100)

for p in g.ax.patches:
    txt = str(p.get_height().round(2)) + '%'
    txt_x = p.get_x() 
    txt_y = p.get_height()
    g.ax.text(txt_x,txt_y,txt)

    plt.savefig("grupos_clases_vivas.png")
```

```python
df_splited = df_titanic.loc[(df_titanic.who == "woman") | (df_titanic.who == "child")]
plot_categorical_relationship_fin(df_splited, "who", "class", relative_freq= True, show_values= True)

plt.savefig("mujeres_nios_vivas.png")
```
